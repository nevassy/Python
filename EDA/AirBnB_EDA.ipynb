### AirBnB - Exploratory Analysis

```python
# Import the libraries
import pandas as pd
```

```python
# Load the dataset
url = 'https://storage.googleapis.com/introduction-to-data-science/airbnb-dataset.csv'
df = pd.read_csv(url)
```

#### Understanding the data

```python
# Display first 5 rows
df.head(5)
```

```python
# Number of rows and columns
df.shape
```

```python
# Name of the columns
df.columns
```

```python
# Types of data
df.dtypes
```

#### Missing values

```python
# Check if there are missing values
df.isnull().sum()
```

```python
# Drop columns id, host_name, last_review

columns_to_drop = ['id', 'host_name', 'last_review']
df.drop(columns_to_drop, axis= "columns", inplace = True)
```

```python
# Check the result 
df.head(5)
```

```python
# Replace missing data in reviews_per_month column

df.fillna({'reviews_per_month': 0}, inplace= True)
# Check the result
df.head(5)
```

```python
# Select the columns: name, neighbourhood_group, price
df[['name', 'neighbourhood_group', 'price']]
```

```python
# Display the first 9 rows of the dataset
df[0:9]
```

```python
# Display the first 9 columns of this dataframe indexed by name, neighbourhood_group, price
df[0:9][['name', 'neighbourhood_group', 'price']]
```

```python
# Determine the departments with price below 100
df[df['price'] < 100]
```

#### Exploratory Data Analysis

```python
# Determine the 10 most reviewed listings

df.nlargest(10, 'number_of_reviews')
```

```python
# Determine what are the different neighourhoods in NYC
df['neighbourhood'].unique()
```

```python
# Determine what are the different neighourhood groups in NYC
df['neighbourhood_group'].unique()
```

```python
# Determine how many departments there are per neighbourhood group
df['neighbourhood_group'].value_counts()
```

```python
# Determine What are the Top 10 neighbourhoods
df['neighbourhood'].value_counts().head(10)
```

```python
# Plot the repartion of the Top 10 neighbourhoods on a histogram
df['neighbourhood'].value_counts().head(10).plot(kind='bar')
```

```python
# Plot the number of department in the neighbourhood groups using seaborn library
import seaborn as sns
sns.countplot(data=df, x= 'neighbourhood_group')
```

```python
# List the different room types
df['room_type'].unique()
```

```python
# Plot the types of rooms per neighbourhood group (using seaborn library)
sns.countplot(data=df, x='neighbourhood_group', hue='room_type')
```

```python
# Determine the distribution of department prices
sns.distplot(df['price'])
```

```python
# Plot the distribution of the departments which have a price below 500
low_price = df[df['price'] <= 500]
sns.distplot(low_price['price'])
```

```python
# Determine the average price of a department (considering departments which have a price below 500)
low_price['price'].mean()
```

```python
# Plot the distribution of department prices based on the Neighbourhood Group (use matplotlib library)
from matplotlib import pyplot as plt
plt.figure(figsize=(15, 8))
sns.violinplot(data=low_price, x='neighbourhood_group', y= 'price')
```

```python
# Plot the listings on a map
low_price.plot(
    kind='scatter',
    x='longitude',
    y='latitude',
    c='price',
    cmap='inferno',
    colorbar=True,
    alpha=0.8,
    figsize=(12,8))
```
