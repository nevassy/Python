##  Working with missing data

###     Identify and handle missing values

```python
# Identify missing values
# Convert "?" to NaN

df.replace("?", np.nan, inplace = True)
df.head(5)
```

### Evaluating for Missing Data

```python
# To detect missing data

missing_data = df.isnull()
missing_data.head(5)
```

### Count missing values in each column

```python
# Count missing values in each column

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

### Binning

Segment the "horsepower" column into 3 bins:
- Convert data to the correct format
- Plot the histogram to see the distribution
- Find 3 bins of equal size bandwidth using Numpy's linspace function
- Build a bin array with a minimum value to a maximum value by using the bandwidth calculated
- Apply the function "cut" to determine what each value the attribute belongs to
- set the number of vehicles in each bin
- Plot the distribution of each bin




```python
# Binning Data in Pandas - "horsepower"
```

```python
# Convert the data to correct format:
df["horsepower"] = df["horsepower"].astype("int")
```

```python
# Plot the histogram to see the distribution of horsepower:

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"], color='aquamarine', edgecolor='black')

# Set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Build an array with a minimum and maximum value

bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
# Set group names
group_names = ['Low', 'Medium', 'High']
```

```python
# Apply the cut function to determine what each value of df['horsepower'] belongs to

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels = group_names, include_lowest=True)
df[['horsepower', 'horsepower-binned']].head(20)
```

```python
# See the number of units in each bin

df['horsepower-binned'].value_counts()
```

```python
# Plot the distribution on each bin

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df['horsepower-binned'].value_counts(), color='aquamarine', edgecolor='black')

# Set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Bins Visualization
# Use histogram to visualize the distribution of bins we created before
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot


# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(df["horsepower"], bins = 3, color='aquamarine', edgecolor='black')

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

### Indicator Variable

##### Indicator Variable for the column "fuel-type"

```python
# Use panda Method 'get_dummies' to assign numerical values to different categories of fuel type
df.columns
```

```python
# Get the indicator variables and assign it to data frame "dummy_variable_1"

dummy_variable_1 = pd.get_dummies(df["fuel-type"], dtype=int)
dummy_variable_1.head(10)
```

```python
# The dataframe, column "fuel-type" now has values for "gas" and "diesel" as 0s and 1s
```

```python
# Change the column names for clarity:
dummy_variable_1.rename(columns={'gas': 'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head(10)
```

```python
# merge data frame "df" and "dummy_variable_1"
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis=1, inplace=True)
```

```python
df.head(10)
```

##### Indicator Variable for the column "aspiration"

```python
# Get the indicator variables and assign it to data frame "dummy_variable_2"

dummy_variable_2 = pd.get_dummies(df["aspiration"], dtype=int)
dummy_variable_2.head(10)
```

```python
# The dataframe, column "aspiration" now has values for "std" and "turbo" as 0s and 1s
```

```python
# Change the column names for clarity:
dummy_variable_2.rename(columns={'std': 'aspiration-std', 'turbo':'aspiration-turbo'}, inplace=True)
dummy_variable_2.head(10)
```

```python
# merge data frame "df" and "dummy_variable_2"
df = pd.concat([df, dummy_variable_2], axis=1)

# drop original column "aspiration" from "df"
df.drop("aspiration", axis=1, inplace=True)
```
